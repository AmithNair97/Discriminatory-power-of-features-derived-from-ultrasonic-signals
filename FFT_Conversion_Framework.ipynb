# Convert WithPerson or WithoutPerson to FFT

```python
# Select from the 17th column to the end

#For With person
adc_data_selected_columns = df_withperson.iloc[:, 16:]# Python uses 0-based indexing

# For Without Person
#adc_data_selected_columns = df_withoutperson.iloc[:, 16:]

#adc_data_selected_columns=adc_data_selected_columns[:11584]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import correlate, hamming

# Convert the pandas DataFrame to a numpy array
adc_array = adc_data_selected_columns.to_numpy()

# Define the sampling frequency
sampling_rate = 1953125  # Use the actual sampling rate

# Initialize lists to store the results
fft_values_list = []
frequency_list = []

# Define a function to compute autocorrelation
def compute_autocorrelation(signal):
    result = correlate(signal, signal, mode='full')
    return result[result.size // 2:]

# Perform FFT for each row
for row in range(adc_array.shape[0]):
    # Compute the autocorrelation
    autocorr_signal = compute_autocorrelation(adc_array[row, :])
    
    # Apply a Hamming window
    window = hamming(len(autocorr_signal))
    windowed_signal = autocorr_signal * window
    
    # Compute the FFT
    fft_result = np.fft.fft(windowed_signal)
    
    # Frequency bins
    freq = np.fft.fftfreq(len(fft_result), d=1/sampling_rate)
    
    # Extract positive frequencies and corresponding FFT values
    positive_freqs = freq[:len(freq) // 2] / 1000  # Convert to kHz
    positive_fft_values = np.abs(fft_result[:len(freq) // 2])
    
    # Compute the Power Spectral Density (PSD)
    psd_values = (positive_fft_values ** 2) / len(windowed_signal)
    
    # Append the results to the lists
    if row == 0:
        frequency_list = positive_freqs  # Store frequency values only once
    fft_values_list.append(psd_values)

# Convert lists to numpy arrays
fft_values = np.array(fft_values_list)

# Create a DataFrame
fft_df = pd.DataFrame(fft_values, columns=frequency_list)

# Define the specific frequency range in kHz
range_min, range_max = 30, 50

# Filter data for the specified range
filtered_columns = [col for col in fft_df.columns if range_min <= col <= range_max]
filtered_data = fft_df[filtered_columns]

# Plot both the time-domain signal and the FFT
plt.figure(figsize=(14, 10))

# Plot the time-domain signal
plt.subplot(2, 1, 1)
plt.plot(np.arange(adc_array.shape[1]) / sampling_rate, adc_array[0, :])  # Plotting the first row as an example
plt.title('Time-Domain Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Plot the FFT for the 30-50 kHz range
plt.subplot(2, 1, 2)
plt.plot(filtered_data.columns, filtered_data.iloc[0, :], label='PSD Value')  # Plotting the first row's PSD values
plt.title('PSD for Frequency Range 30-50 kHz')
plt.xlabel('Frequency (kHz)')
plt.ylabel('Magnitude')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
