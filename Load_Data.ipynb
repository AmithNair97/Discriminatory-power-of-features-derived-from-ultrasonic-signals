```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

# Load the DataFrame values and columns
values = np.load('with_person_filtered_data_values.npy')
columns = np.load('with_person_filtered_data_columns.npy', allow_pickle=True)

# Convert the loaded values back to a DataFrame with the original column names
loaded_df = pd.DataFrame(values, columns=columns)

# Load the second DataFrame values and columns
values1 = np.load('without_person_filtered_data_values.npy')
columns1 = np.load('without_person_filtered_data_columns.npy', allow_pickle=True)

# Convert the loaded values back to a DataFrame with the original column names
loaded_df1 = pd.DataFrame(values1, columns=columns1)

# Concatenate the DataFrames
df = pd.concat([loaded_df, loaded_df1], ignore_index=True)

# Replace NaN values with 0
df=df.dropna()

# Display the resulting DataFrame
df
```

# Model Building

```python
# Import every packages for EDA

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# plot to appear inside notebook
%matplotlib inline 

# Models from Sci-kit learn for models

from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier

# Model evaluation 

from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV, GridSearchCV
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import precision_score,recall_score,f1_score
from sklearn.metrics import roc_curve
```

```python
X = df.drop('Person Presence',axis=1)
X
```

```python
Y = df['Person Presence']
Y
```

```python
# Split the data 

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
# Lets try to fit and score the models using  ML models 
np.random.seed(42)
models ={
    #'SVC': SVC(probability=True),
    'KNN Classifier':KNeighborsClassifier(),
    'Random Forest Classifier': RandomForestClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Naive Bayes': GaussianNB(),
    'XGBoost Classifier': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
}

np.random.seed(42)

model_scores={}

for name,model in models.items():
    model.fit(x_train,y_train)
    model_scores[name] = model.score(x_test,y_test)

model_scores
```

```python
# Assuming model_scores is already defined
model_comparison = pd.DataFrame(model_scores, index=['accuracy'])

# Increase the figure size for better clarity
plt.figure(figsize=(10, 6))

# Plot the bar chart with adjusted rotation for better readability
model_comparison.T.plot.bar(color='lightblue', legend=False)

# Adjust x-axis labels for better clarity
plt.xticks(rotation=15, fontsize=12, ha='right')

# Set y-axis label
plt.ylabel('Accuracy', fontsize=14)

# Set x-axis label
plt.xlabel('Models', fontsize=14)

# Set chart title
plt.title('Model Accuracy Comparison', fontsize=16)

# Display the chart
plt.show()
```

# KNN


# Classification Report 

```python
y_preds_KNN=models['KNN Classifier'].predict(x_test)
y_preds_KNN
pd.DataFrame(data=classification_report(y_test,y_preds_KNN,output_dict=True))

```

# ROC Curve

```python

# Predict probabilities for KNN
y_preds_KNN = models['KNN Classifier'].predict_proba(x_test)[:, 1]

# Convert probabilities to binary class labels
y_preds_binary_KNN = (y_preds_KNN > 0.5).astype(int)

# Compute ROC curve
fpr_KNN, tpr_KNN, thresholds_KNN = roc_curve(y_test, y_preds_binary_KNN)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_KNN, tpr_KNN, label='ROC Curve (KNN)')
plt.plot([0, 1], [0, 1], linestyle='--', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve - KNN')
plt.legend()
plt.show()
```

# Confusion matrix

```python

y_preds_binary_KNN = (y_preds_KNN > 0.5).astype(int)
cm_KNN = confusion_matrix(y_test, y_preds_binary_KNN)
print(cm_KNN)

# Plot confusion matrix for KNN
fig, ax = plt.subplots(figsize=(8, 6))
cax = ax.matshow(cm_KNN, cmap='Blues')
plt.title('Confusion Matrix - KNN')
fig.colorbar(cax)
plt.xlabel('True label')
plt.ylabel('Predicted label')

for (i, j), val in np.ndenumerate(cm_KNN):
    ax.text(j, i, f'{val}', ha='center', va='center')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming the models are already trained
feature_importances = {
    'Decision Tree': models['Decision Tree'].feature_importances_,
    'Random Forest': models['Random Forest Classifier'].feature_importances_,
    'XGBoost': models['XGBoost Classifier'].feature_importances_,
    'Logistic Regression': models['Logistic Regression'].coef_[0],  # Assuming binary classification
    'Naive Bayes': np.mean(models['Naive Bayes'].feature_log_prob_, axis=0)
}

# Create dictionaries for feature importances
# Assuming 'x_train.columns' contains the feature names
decision_tree_dict = dict(zip(x_train.columns, feature_importances['Decision Tree']))
random_forest_dict = dict(zip(x_train.columns, feature_importances['Random Forest']))
xgboost_dict = dict(zip(x_train.columns, feature_importances['XGBoost']))
logreg_dict = dict(zip(x_train.columns, feature_importances['Logistic Regression']))
naive_bayes_dict = dict(zip(x_train.columns, feature_importances['Naive Bayes']))

# Create DataFrames from the dictionaries
decision_tree_df = pd.DataFrame(data=decision_tree_dict, index=[0]).T
random_forest_df = pd.DataFrame(data=random_forest_dict, index=[0]).T
xgboost_df = pd.DataFrame(data=xgboost_dict, index=[0]).T
logreg_df = pd.DataFrame(data=logreg_dict, index=[0]).T
naive_bayes_df = pd.DataFrame(data=naive_bayes_dict, index=[0]).T

# Name the columns appropriately
decision_tree_df.columns = ['Decision Tree Importance Value']
random_forest_df.columns = ['Random Forest Importance Value']
xgboost_df.columns = ['XGBoost Importance Value']
logreg_df.columns = ['Logistic Regression Importance Value']
naive_bayes_df.columns = ['Naive Bayes Importance Value']

# Calculate the absolute values
decision_tree_df['Absolute Decision Tree Importance Value'] = decision_tree_df['Decision Tree Importance Value'].abs()
random_forest_df['Absolute Random Forest Importance Value'] = random_forest_df['Random Forest Importance Value'].abs()
xgboost_df['Absolute XGBoost Importance Value'] = xgboost_df['XGBoost Importance Value'].abs()
logreg_df['Absolute Logistic Regression Importance Value'] = logreg_df['Logistic Regression Importance Value'].abs()
naive_bayes_df['Absolute Naive Bayes Importance Value'] = naive_bayes_df['Naive Bayes Importance Value'].abs()

# Extract the frequency range from the feature names
# Assuming feature names are like 'feature_30', 'feature_31', ..., 'feature_50'
decision_tree_df['Frequency Range'] = decision_tree_df.index.to_series().astype(str).str.extract(r'(\d+)', expand=False).astype(int)
random_forest_df['Frequency Range'] = random_forest_df.index.to_series().astype(str).str.extract(r'(\d+)', expand=False).astype(int)
xgboost_df['Frequency Range'] = xgboost_df.index.to_series().astype(str).str.extract(r'(\d+)', expand=False).astype(int)
logreg_df['Frequency Range'] = logreg_df.index.to_series().astype(str).str.extract(r'(\d+)', expand=False).astype(int)
naive_bayes_df['Frequency Range'] = naive_bayes_df.index.to_series().astype(str).str.extract(r'(\d+)', expand=False).astype(int)

# Group by the frequency range and compute the sum of absolute importance values
decision_tree_importance = decision_tree_df.groupby('Frequency Range')['Absolute Decision Tree Importance Value'].sum().sort_index()
random_forest_importance = random_forest_df.groupby('Frequency Range')['Absolute Random Forest Importance Value'].sum().sort_index()
xgboost_importance = xgboost_df.groupby('Frequency Range')['Absolute XGBoost Importance Value'].sum().sort_index()
logreg_importance = logreg_df.groupby('Frequency Range')['Absolute Logistic Regression Importance Value'].sum().sort_index()
naive_bayes_importance = naive_bayes_df.groupby('Frequency Range')['Absolute Naive Bayes Importance Value'].sum().sort_index()

# Normalize the importance values
decision_tree_importance_normalized = (decision_tree_importance - decision_tree_importance.min()) / (decision_tree_importance.max() - decision_tree_importance.min())
random_forest_importance_normalized = (random_forest_importance - random_forest_importance.min()) / (random_forest_importance.max() - random_forest_importance.min())
xgboost_importance_normalized = (xgboost_importance - xgboost_importance.min()) / (xgboost_importance.max() - xgboost_importance.min())
logreg_importance_normalized = (logreg_importance - logreg_importance.min()) / (logreg_importance.max() - logreg_importance.min())
naive_bayes_importance_normalized = (naive_bayes_importance - naive_bayes_importance.min()) / (naive_bayes_importance.max() - naive_bayes_importance.min())

# Combine the normalized importance values into a single DataFrame for plotting
combined_df_normalized = pd.DataFrame({
    'Decision Tree': decision_tree_importance_normalized,
    'Random Forest': random_forest_importance_normalized,
    'XGBoost': xgboost_importance_normalized,
    'Logistic Regression': logreg_importance_normalized,
    'Naive Bayes': naive_bayes_importance_normalized
})

# Plot the line chart for the combined normalized feature importance
ax = combined_df_normalized.plot(
    title='Normalized Importance of Frequency Ranges',
    figsize=(12, 8),
    marker='o'
)
ax.set_xlabel('Frequency Range')
ax.set_ylabel('Normalized Sum of Absolute Importance Values')

plt.show()

# Print the combined normalized importance values in a readable format
print("Normalized importance of frequency ranges (sum of absolute importance values) across models:")
print(combined_df_normalized)

# Display the DataFrame for reference
combined_df_normalized.reset_index(inplace=True)
combined_df_normalized
```
